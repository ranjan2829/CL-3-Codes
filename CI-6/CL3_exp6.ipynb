```python
import numpy as np

# Define the problem (e.g., optimization problem)
def objective_function(x):
    # Example: Sphere function (minimize the sum of squares)
    return np.sum(x**2)

# Initialize parameters
population_size = 100
num_generations = 50
num_clones = 10
mutation_rate = 0.1
dimension = 5  # Dimensionality of the problem

# Initialize the population of antibodies (solutions)
population = np.random.uniform(-10, 10, (population_size, dimension))

# Clonal Selection Algorithm
for generation in range(num_generations):
    # Evaluate the affinity (fitness) of each antibody
    fitness = np.array([objective_function(individual) for individual in population])

    # Select the best antibodies based on their affinity
    selected_indices = np.argsort(fitness)[:population_size // 2]
    selected_population = population[selected_indices]

    # Clone the selected antibodies
    clones = np.repeat(selected_population, num_clones, axis=0)

    # Mutate the cloned antibodies
    mutations = np.random.normal(0, mutation_rate, clones.shape)
    mutated_clones = clones + mutations

    # Evaluate the mutated clones
    mutated_fitness = np.array([objective_function(individual) for individual in mutated_clones])

    # Select the best antibodies from the mutated clones
    best_indices = np.argsort(mutated_fitness)[:population_size // 2]
    best_clones = mutated_clones[best_indices]

    # Form the new population
    population = np.vstack((best_clones, np.random.uniform(-10, 10, (population_size // 2, dimension))))

    # Print the best solution found in this generation
    best_solution = population[np.argmin(fitness)]
    print(f"Generation {generation}: Best Solution = {best_solution}, Fitness = {np.min(fitness)}")

# Final best solution
best_solution = population[np.argmin(fitness)]
print(f"Best Solution Found: {best_solution}, Fitness = {np.min(fitness)}")
```
